### Solution & Pseudocode

image.png

```python
arr = [80, 85, 90, 95, 20, 30, 35, 50, 60, 65, 67, 75]
# arr = [1, 2, 3, 4, 5, 6, 7, 8, 9]
# arr = [9, 1, 2, 3, 4, 5, 6, 7, 8]
def find_k(arr):
  '''
  Function này tìm được smallest element trong sorted rotated array(tức là tìm cái điểm hẫng)

  Cách chặt nhị phân mà không cần lưu thêm 1 biến để lưu đáp án:(Với các bài tìm số trong dãy tman tchat)
  +/ Điều kiện kết thúc không còn là l <= r nữa mà sẽ thay đổi thành l < r or l + 2 <= r như code
  +/ Kết quả luôn nằm ở biến hi, so sánh A[mid] với A[hi]
    - Nếu A[mid] tốt hơn A[hi] -> hi = mid(tức là hi luôn dịch theo cái tốt nhất, tốt hơn thì update hi)
    - Ngước lại lo = mid (nếu không thay đổi được tốt hơn thì update lo)
    Ví dụ ở bài này "tốt hơn" : tương đương với nhỏ hơn
  '''
  lo, hi = 0, len(arr) - 1
  while lo + 2 <= hi:
    mid = (lo + hi) // 2
    if arr[mid] < arr[hi]:
      hi = mid
    else:
      lo = mid
  # print(lo, hi)
  return hi

def check_x(arr, x):
  # check if x is in array or not
  lo, hi = 0, len(arr) - 1
  # print(arr)
  while lo + 2 <= hi:
    # print(lo, hi)
    mid = (lo + hi) // 2
    if arr[mid] >= x:
      hi = mid
    else:
      lo = mid
  return arr[hi] == x

k = find_k(arr)
x = 70
print(check_x(arr[:k], x) or check_x(arr[k:], x))

```
