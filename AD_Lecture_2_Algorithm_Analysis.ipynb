### Solution & Pseudocode

image.png

```python
arr = [80, 85, 90, 95, 20, 30, 35, 50, 60, 65, 67, 75]
# arr = [1, 2, 3, 4, 5, 6, 7, 8, 9]
# arr = [9, 1, 2, 3, 4, 5, 6, 7, 8]
def find_k(arr):
  '''
  Function ny tm c smallest element trong sorted rotated array(tc l tm ci im hng)

  Cch cht nh phn m khng cn lu thm 1 bin  lu p n:(Vi cc bi tm s trong dy tman tchat)
  +/ iu kin kt thc khng cn l l <= r na m s thay i thnh l < r or l + 2 <= r nh code
  +/ Kt qu lun nm  bin hi, so snh A[mid] vi A[hi]
    - Nu A[mid] tt hn A[hi] -> hi = mid(tc l hi lun dch theo ci tt nht, tt hn th update hi)
    - Ngc li lo = mid (nu khng thay i c tt hn th update lo)
    V d  bi ny "tt hn" : tng ng vi nh hn
  '''
  lo, hi = 0, len(arr) - 1
  while lo + 2 <= hi:
    mid = (lo + hi) // 2
    if arr[mid] < arr[hi]:
      hi = mid
    else:
      lo = mid
  # print(lo, hi)
  return hi

def check_x(arr, x):
  # check if x is in array or not
  lo, hi = 0, len(arr) - 1
  # print(arr)
  while lo + 2 <= hi:
    # print(lo, hi)
    mid = (lo + hi) // 2
    if arr[mid] >= x:
      hi = mid
    else:
      lo = mid
  return arr[hi] == x

k = find_k(arr)
x = 70
print(check_x(arr[:k], x) or check_x(arr[k:], x))

```
